```python
import pandas as pd
import numpy as np
import os
import warnings
import lightgbm as lgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from IPython.display import display, HTML
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8')
print("✅ Libraries loaded.")
```

```python
DATASET_DIR = r'C:\Users\cheku\OneDrive\Documents\network_project'

def get_label_from_filename(filename):
    name = filename.lower()
    if 'syn' in name: return "SYN_Flood"
    elif 'icmp' in name: return "ICMP_Flood"
    elif 'udp' in name: return "UDP_Flood"
    elif 'slow' in name or 'http' in name: return "Slowloris"
    elif 'normal' in name or 'benign' in name: return "Normal"
    else: return None

all_files = {}
training_data = []

print("🔍 Loading datasets...")

for filename in sorted(os.listdir(DATASET_DIR)):
    if not filename.endswith('.csv'): continue
    df = pd.read_csv(os.path.join(DATASET_DIR, filename))
    if len(df) == 0: continue

    true_label = get_label_from_filename(filename)
    if true_label is not None:
        df_labeled = df.copy()
        df_labeled['label'] = true_label
        training_data.append(df_labeled)
        print(f"✅ {filename} → {true_label} ({len(df)} packets)")
    else:
        print(f"🕵️ {filename} → Unknown ({len(df)} packets)")
    all_files[filename] = df

combined = pd.concat(training_data, ignore_index=True)
FEATURES = ['length', 'protocol_type', 'syn_flag', 'ack_flag', 
            'icmp_payload_size', 'dst_port', 'is_http_traffic', 'is_partial_http']
X = combined[FEATURES].fillna(0)
y = combined['label']
print(f"\n📊 Training data: {len(combined)} samples | Classes: {sorted(y.unique())}")
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
y.value_counts().plot(kind='bar', color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7'])
plt.title("Class Distribution"); plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.heatmap(X.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation")
plt.tight_layout(); plt.show()

display(HTML("<h3>🔍 Feature Stats by Class</h3>"))
display(combined.groupby('label')[FEATURES].agg(['mean', 'std']).round(2))
```

```python
import joblib
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

rf_model = RandomForestClassifier(n_estimators=200, class_weight='balanced', random_state=42)
rf_model.fit(X_train_scaled, y_train)
rf_acc = rf_model.score(X_test_scaled, y_test)

joblib.dump(rf_model, "model.pkl")
joblib.dump(scaler, "scaler.pkl")          # correct name
# joblib.dump(label_encoder, "encoder.pkl")  # only if exists

feature_names = FEATURES   # or list(X.columns)
joblib.dump(feature_names, "features.pkl")

lgb_model = lgb.LGBMClassifier(n_estimators=200, class_weight='balanced', random_state=42, verbosity=-1)
lgb_model.fit(X_train_scaled, y_train)
lgb_acc = lgb_model.score(X_test_scaled, y_test)

print(f"✅ RF Accuracy: {rf_acc:.1%} | LightGBM: {lgb_acc:.1%}")

plt.figure(figsize=(8, 4))
imp = rf_model.feature_importances_
idx = np.argsort(imp)[::-1]
plt.bar(range(len(FEATURES)), imp[idx], color='#6C5CE7')
plt.xticks(range(len(FEATURES)), [FEATURES[i] for i in idx], rotation=45)
plt.title("Feature Importance"); plt.tight_layout(); plt.show()
```

```python
def get_firewall_commands_all_os(ip="127.0.0.1"):
    ip_safe = ip.replace(".", "_")
    return {
        'SYN_Flood': {'name':'TCP SYN Flood','color':'#FF4757',
            'windows':f'netsh advfirewall firewall add rule name="BLOCK_SYN_{ip_safe}" dir=in action=block remoteip={ip}',
            'linux':f'sudo iptables -A INPUT -s {ip} -p tcp --syn -j DROP',
            'macos':f'sudo pfctl -t blocked_ips -T add {ip} && echo "block in from {ip}" | sudo pfctl -f -'},
        'ICMP_Flood': {'name':'ICMP Flood','color':'#2ED573',
            'windows':f'netsh advfirewall firewall add rule name="BLOCK_ICMP_{ip_safe}" protocol=icmpv4 dir=in action=block remoteip={ip}',
            'linux':f'sudo iptables -A INPUT -s {ip} -p icmp -j DROP',
            'macos':f'sudo pfctl -t blocked_ips -T add {ip} && echo "block in proto icmp from {ip}" | sudo pfctl -f -'},
        'UDP_Flood': {'name':'UDP Flood','color':'#FFA502',
            'windows':f'netsh advfirewall firewall add rule name="BLOCK_UDP_{ip_safe}" protocol=UDP dir=in action=block remoteip={ip}',
            'linux':f'sudo iptables -A INPUT -s {ip} -p udp -j DROP',
            'macos':f'sudo pfctl -t blocked_ips -T add {ip} && echo "block in proto udp from {ip}" | sudo pfctl -f -'},
        'Slowloris': {'name':'Slowloris','color':'#6C5CE7',
            'windows':'# Configure web server: Timeout 30, MaxRequestWorkers 100',
            'linux':'# Apache: Timeout 30; Nginx: client_header_timeout 30s;',
            'macos':'# Same as Linux – adjust web server config'},
        'Normal': {'name':'Normal Traffic','color':'#218c74',
            'windows':'# No blocking needed',
            'linux':'# No blocking needed',
            'macos':'# No blocking needed'}
    }
```

```python
print("\n🔍 ANALYZING ALL FILES (with mitigation commands)...\n")

for filename in sorted(all_files.keys()):
    df = all_files[filename]
    if len(df) == 0:
        continue

    X_file = df[FEATURES].fillna(0)
    X_scaled = scaler.transform(X_file)
    preds = rf_model.predict(X_scaled)
    confs = rf_model.predict_proba(X_scaled).max(axis=1)
    
    pred_counter = Counter(preds)
    dominant = pred_counter.most_common(1)[0][0]
    pct = pred_counter[dominant] / len(preds) * 100
    avg_conf = confs.mean()
    
    # Get commands for detected attack
    cmds = get_firewall_commands_all_os("127.0.0.1")[dominant]
    is_unknown = filename.startswith('unknown')
    
    # Format firewall commands for display
    cmd_block = f"""[Windows]
{cmds['windows']}

[Linux]
{cmds['linux']}

[macOS]
{cmds['macos']}"""
    
    display(HTML(f'''
    <div style="background: linear-gradient(135deg, #f8f9fa 0%, #e9ecef 100%); 
                border-left:6px solid {cmds['color']}; padding:16px; margin:12px 0; border-radius:0 8px 8px 0;
                {'box-shadow: 0 0 12px rgba(0,0,0,0.15);' if is_unknown else ''}">
        <h4 style="margin:0; color:#495057;">{'🕵️ UNKNOWN FILE' if is_unknown else '📊 Known File'}: {filename}</h4>
        <p style="margin:8px 0; font-size:0.95em; font-weight: {'bold' if is_unknown else 'normal'};">
            <strong>Detected Attack:</strong> {dominant} ({pct:.1f}% of packets)<br>
            <strong>Total Packets:</strong> {len(preds)}<br>
            <strong>Average Confidence:</strong> {avg_conf:.1%}
        </p>
        <details style="margin-top:10px;">
            <summary><strong>🛡️ Mitigation Commands (Click to Expand)</strong></summary>
            <pre style="background:#2d3436; color:#00b894; padding:12px; border-radius:5px; font-size:12px; white-space:pre-wrap; margin-top:8px;">{cmd_block}</pre>
        </details>
    </div>'''))

print("\n✅ Analysis complete! Ready to defend.")
```
